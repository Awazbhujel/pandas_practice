**Probability Ratio Encoding**
1. Probability of Survived based on Cabin -- Categorical Features
2. if i know the probability of Survived then i can find the probability of not Survived -- not Survived(1-Survived)
3. finding the probability ratio by  dividing survived  with not survived columns
4. Making the dictionary to map cabin with probability ratio
5. Replace with Categorical Feature

```python
#reading the dataset
df=pd.read_csv('/content/train.csv',usecols=['Cabin','Survived'])
df.head()
```

```python
#filling the nan values with "Missing"
df["Cabin"].fillna("Missing",inplace=True)
df.head()
```

```python
#taking only the cabin number
df["Cabin"]=df["Cabin"].astype(str).str[0]
df.head()
```

```python
#finding the probability of survived columns with respect to cabin number
prob_df=df.groupby(["Cabin"])["Survived"].mean()
```

```python
#making the dataframe of prob_df
prob_df = pd.DataFrame(prob_df)
prob_df
```

```python
#adding the columns dead by subtracting 1 with survived mean 
prob_df["Died"] = 1-prob_df["Survived"]
```

```python
prob_df.head(10)
```

```python
#probability ratio
prob_df["Survived"]/prob_df["Died"]
```

```python
#making probability ratio by  dividing survived columns with died columns 
prob_df["Probability_ratio"] = prob_df["Survived"]/prob_df["Died"]
```

```python
prob_df.head(10)
```

```python
#making the dictionary to map with Cabin columns
prob_encoded=prob_df["Probability_ratio"].to_dict()
```

```python
df["Cabin_encoded"]= df["Cabin"].map(prob_encoded)
```
